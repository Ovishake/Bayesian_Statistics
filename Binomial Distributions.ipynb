```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 20
p= 0.5
x = np.arange(0,21)

plt.plot(x, stats.binom.pmf(x,20,0.5), 'bo', ms=8)
plt.vlines(x, 0, stats.binom.pmf(x,n,p), colors="b", lw=5, alpha=0.5)
```

```python
plt.plot(x, stats.binom.pmf(x,20,0.8), 'bo', ms=8)
plt.vlines(x, 0, stats.binom.pmf(x,n,0.8), colors="b", lw=5, alpha=0.5)
```

```python
plt.plot(x, stats.binom.pmf(x,20,0.2), 'bo', ms=8)
plt.vlines(x, 0, stats.binom.pmf(x,n,0.2), colors="b", lw=5, alpha=0.5)
```

```python
x = np.arange(0,21)
plt.plot(x, np.ones(21)/21, 'bo', ms=8) # here the division happens
plt.vlines(x,0, np.ones(21)/21, colors="b", lw=5, alpha=0.5)
plt.ylim(0, 0.2)
```

```python
mu=0.5
sigma=0.1

x = np.linspace(0,1,100)
plt.plot(x, stats.norm.cdf(x, mu, sigma), 'b-', lw=5, alpha=0.6)
plt.ylim(0,1)
```

```python
mu=0.5
sigma=0.1

x = np.linspace(0,1,100)
plt.plot(x, stats.norm.pdf(x, mu, sigma), 'b-', lw=5, alpha=0.6)
plt.ylim(0,5)
```

a, b are defined over the range of the standard normal
a and b are shape parameters

```python
# standard normal
a, b = -5, 5

x = np.linspace(stats.truncnorm.ppf(0.01, a, b),
                stats.truncnorm.ppf(0.99, a, b), 100)

plt.plot(x, stats.truncnorm.pdf(x, a, b),
       'b-', lw=5, alpha=0.6, label='truncnorm pdf')
```
